### Data Cleaning and Processing

Get total medals only

```python
total_medal_df = medal_df.copy()
total_medal_df=total_medal_df.drop(['Rank', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'RankbyTotal', 'NOCCode'], axis=1)
total_medal_df.head()
```

Rename Olympics countries names with those from World Bank

```python
total_medal_df['Team/NOC'] = total_medal_df['Team/NOC'].replace(["Bahamas","Chinese Taipei","CÃ´te d'Ivoire","Egypt","Great Britain","Hong Kong, China","Islamic Republic of Iran","Kyrgyzstan","People's Republic of China","ROC","Republic of Korea","Republic of Moldova","Slovakia","United States of America","Venezuela"],["Bahamas, The","Taiwan","Cote d'Ivoire","Egypt, Arab Rep.","United Kingdom","Hong Kong SAR, China","Iran, Islamic Rep.","Kyrgyz Republic","China","Russian Federation","Korea, Rep.","Moldova","Slovak Republic","United States","Venezuela, RB"])
total_medal_df
```

```python
total_medal_df.columns=['Country Name','Total Medals']
total_medal_df
```

Get 2020 Population and Taiwan's into it

```python
twenty_population_df = population_df.copy()
twenty_population_df = twenty_population_df[['Country Name', '2020']]
twenty_population_df.head()
```

```python
new_row = {'Country Name':'Taiwan', '2020':23561236.0}
twenty_population_df = twenty_population_df.append(new_row, ignore_index=True)
```

```python
merge_pop_medal_df = pd.merge(total_medal_df, twenty_population_df, on=["Country Name"])
merge_pop_medal_df
```

```python
merge_pop_medal_df['Population Per Medal'] = merge_pop_medal_df['2020']/merge_pop_medal_df['Total Medals']
merge_pop_medal_df
```

```python
merge_pop_medal_df.to_csv('merge_pop_medal_df.csv',index=False)
```
